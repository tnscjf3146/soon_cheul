## 데이터 분석도구 : Numpy

## Numpy란? 왜 필요한가?

**Numpy?**
- 대규모 수치 데이터를 빠르게 처리하기 위한 라이브러리
- 핵심 자료구조는 **ndarray (N-dimensional array, N 차원 배열)**
- 데이터 과학, 머신러닝, 과학 계산 등 거의 모든 Python 데이터 작업의 기반이 된다.

### numpy가 필요한 이유
python의 기본 리스트로도 숫자 데이터를 다룰 수 있으나, 한계가 존재


```python
# numpy 사용을 위한 임포트
import numpy as np
```

```python
numbers = [1, 2, 3, 4, 5]
result = []
for num in numbers:
    result.append(num * 2) # 하나씩 반복해야 하기때문에 느림

# ✅ NumPy 사용
numbers = np.array([1, 2, 3, 4, 5])
result = numbers * 2 # 한 번에 처리 빠름
```
